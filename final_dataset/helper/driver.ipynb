```python
# dfile -> data from db
# pfile -> data from pdf

def getData(dfile,pfile):
    data = pickle.load(open(dfile,'rb'))
    print("DB Shape - ",data.shape)
    
    df = pickle.load(open(pfile,'rb'))
    print("PDF Shape - ",df.shape)
    
    one =data['udise_sch_code'].to_list()
    two =df['Code'].to_list()
    two = [int(t) for t in two]
    
    intersection = sorted(set(one).intersection(set(two)))
    
    print("No. of codes in (1)DB - ", len(set(one)), " ,(2)PDF - ", len(set(two)))
    print("No. of codes NOT in PDF - ", len(set(one)-set(two)))
    print("No. of codes NOT in DB - ", len(set(two)-set(one)))
    print("No. of codes in BOTH - ", len(intersection))
    print("\n")
    
    return data, df, intersection

dfile_head = "Excel_Pckls/"
pfile_head = "pckls/"

dist = "ANANTAPUR"
adata, adf, AI = getData(dfile_head+dist, pfile_head+dist)

dist = "CHITTOOR"
cdata, cdf, CI = getData(dfile_head+dist, pfile_head+dist)

dist = "GUNTUR"
gdata, gdf, GI = getData(dfile_head+dist, pfile_head+dist)
```

# Check for duplicates

```python
# Derive code_name pairs and check for duplicates
def getAllPairs():
    code_name = []
    data = [adata, cdata, gdata]
    ids = [AI, CI, GI]

    for df,i in zip(data, ids):
        temp = df.loc[df['udise_sch_code'].isin(i), ['udise_sch_code', 'school_name']]
        temp = temp.values.tolist()
        code_name =code_name + temp
    
    return code_name

# Checks if there are multiple codes linked to one school name
def nameTOcode(pairs):
    pairs = sorted(pairs, key=lambda p:p[1])

    dups = {}
    for pair in pairs:
        code =pair[0]
        name =pair[1]

        if name in dups:
            dups[name].append(code)
        else:
            dups[name]=[code]
            
    dups = sorted(dups.items(), key=lambda d:len(d[1]), reverse=True)
    return dups

# Checks if there are multiple names linked to one udise code
def codeTOname(pairs):
    pairs = sorted(pairs, key=lambda p:p[1])

    dups = {}
    for pair in pairs:
        code =pair[0]
        name =pair[1]

        if code in dups:
            if name not in dups[code]:   #to remove duplicate
                dups[code].append(name)
        else:
            dups[code]=[name]
            
    dups = sorted(dups.items(), key=lambda d:len(d[1]), reverse=True)
    return dups

# Checks if there are multiple pairs of [code, name]
def checkDuplicates(pairs):
    pairs = sorted(pairs, key=lambda p:p[1])
    
    dups ={}
    for pair in pairs:
        if str(pair) in dups:
            dups[str(pair)]+=1
        else:
            dups[str(pair)]=1
        
    dups = sorted(dups.items(),key=lambda d:d[1], reverse=True)
    return dups


pairs = getAllPairs()
# dups =checkDuplicates(pairs)
dups =codeTOname(pairs)
```

```python
pairs =getAllPairs()

namecode = nameTOcode(pairs)
codename = codeTOname(pairs)
dupPairs = checkDuplicates(pairs)
```

```python
def extractClean(namecode, dupPairs):
    allIds = set(AI+GI+CI)
    codes =[]
    
    for row in namecode:
        tmp = row[1]
        if len(tmp) > 1:
            codes +=tmp
    
    for row in dupPairs:
        if row[1] > 1:
            codes.append(int(row[0].strip('"[]').split(',')[0]))
            

    return allIds-set(codes)
   
def prune(simpleCodes, AI):
    ai =[]
    for code in simpleCodes:
        if code in AI:
            ai.append(code)

    return ai
```

```python
simpleCodes = extractClean(namecode, dupPairs)
ai =prune(simpleCodes, AI)
```

```python
'''
Analyse the school names
    - how many start with [mpups, mps, zphs, kgbv...etc]
'''

def getAllNames():
    names = []
    data = [adata, cdata, gdata]
    ids = [AI, CI, GI]

    for df,i in zip(data, ids):
        temp = df.loc[df['udise_sch_code'].isin(i), 'school_name'].drop_duplicates()
        temp = temp.values.tolist()
        print(len(temp))
        names += temp
    
    return names

names =getAllNames()
len(names)
```

```python
def getOther(pairs):
    pairs = sorted(pairs, key=lambda p:p[1])
    sp =['MPPS', 'MPUPS', 'ZPPS', 'ZPHS', 'KGBV']
    other = []

    for pair in pairs:
        name = pair[1]
        if list(filter(name.startswith, sp))==[]:
            other.append(pair)
        
    return other

other = getOther(pairs)
other[:10]
```

```python
fout =open('other.txt', 'w')
for pair in other:
    line = str(pair[0])+" -|- "+pair[1]+"\n"
    fout.write(line)
    
fout.close()
```
